```python
url = "http://news.google.com/news?q=covid-19&hl=en-US&sort=date&gl=US&num=100&output=rss"
texts = []
documentDict = dict()
class Document:
    
#     title
#     url
#     content
#     PublishedDate
    
    def __init__(self, title, url, content, PublishedDate):
        self.title = title
        self.url = url
        self.content = content
        self.PublishedDate = PublishedDate
    
    
class ParseFeed():
    

    def __init__(self, url):
        self.feed_url = url
        
    def clean(self, html):
        '''
        Get the text from html and do some cleaning
        '''
        soup = BeautifulSoup(html)
        text = soup.get_text()
        text = text.replace('\xa0', ' ')
        text = text.replace('\xa0', ' ')
        text = text.replace('\n', '')
        text = text.replace('\r', '')
        text = text.replace('\t', '')
        out = re.sub('[^\w\d\s]+', '', text)
        return text

    def parse(self):
        '''
        Parse the URL, and print all the details of the news 
        '''
        feeds = feedparser.parse(self.feed_url).entries
        id = 0
        for f in feeds:
            url = f.get("link", "")
            try: response = urllib.request.urlopen(url)

            except HTTPError as e:
                     print("Something bad happened")
            except URLError as e:
                     print("URLError happened")
           
            except urllib3.exceptions.SSLError as e:
                     print("ssl ERROR")
                
            
            else:
                id = id + 1
                text = self.clean(response)
                d = Document(f.get("title", ""),url,text,f.get("published", ""))
                documentDict[id] = d
                
                pprint({
                    'Description': self.clean(f.get("description", "")),
                    'Published Date': f.get("published", ""),
                    'Title': f.get("title", ""),
                    'Url': f.get("link", ""),
                    'text': text

                })
        
     
feed = ParseFeed(url)
feed.parse()
```

```python
sentimentDict = dict()

def fetch_results(docId):    
    value = documentDict[docId]
    text = value.content
   # create list of sentence tokens
    sentences = text.split(".")
    count = Counter()
    sentimentScore = 0

    for sentence in sentences:
        try: result = model.predict(sentence)
        except AttributeError as e:
                             print("AttributeError") 
        else:  
                count[result[0]] += 1
                    
    sentimentDict[docId] = max(count.items(), key=lambda x: x[1])
```

```python
finalResult = []
for key in sentimentDict:
    docTuple = sentimentDict[key]
    if docTuple[0] == 'positive':
         docVal = documentDict[key]
         finalResult.append(docVal)
         pprint(pprint({'Title': docVal.title,'Url':docVal.url,'Content':docVal.content}))
```

```python
import json 
import csv 
  
  
# now we will open a file for writing 
data_file = open('PositiveSentimentCovidNews.csv', 'w', encoding="utf8") 
# create the csv writer object 
csv_writer = csv.writer(data_file) 
  
# Counter variable used for writing  
# headers to the CSV file 
count = 0

for doc in finalResult: 
    #convert to JSON string
    jsonStr = json.dumps(doc.__dict__)
    #Convert to JSON object
    docJson = json.loads(jsonStr)

    if count == 0:         
        # Writing headers of CSV file 
        header = docJson.keys()
        csv_writer.writerow(header) 
        count += 1
        
    # Writing data of CSV file 
    csv_writer.writerow(docJson.values()) 
  
data_file.close() 
```
